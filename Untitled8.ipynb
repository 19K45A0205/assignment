5. Unique Sort problem
Take a string as input that accepts a comma separated sequence of words
as input and prints the unique words in sorted form (alphanumerically).
*Input*: orange, white, red, cyan, green, magenta, cyan, pink, white


```
L=["orange","white","red","cyan","green","magenta","cyan","pink","white"]
print("the original list is:",str(L))
L=list(set(L))
print("the list after removing the duplicates is:",str(L))
L.sort()
print(L)
```

9. Sort 3 Integers
Given three integers (given through user input), sort the numbers using
|min| and |max| functions.

```
x=int(input("enter the first number:"))
y=int(input("enter the second number:"))
z=int(input("enter the third number:"))
a1=min(x,y,z)
a3=max(x,y,z)
a2=(x + y + z)-a1-a3
print("the numbers in sorted form are:",a1,a2,a3)
```

11. Compute product of a list of numbers [45 ,3,2,89,72,1,10,7]
Output: 121111200

```
from functools import reduce
number_list=[45 ,3,2,89,72,1,10,7]
numbers_product=reduce((lambda x,y: x*y),number_list)
print("the product of numbers in the list is:",numbers_product)
```
